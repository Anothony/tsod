# Seasonally varying range detection

```python
import pandas as pd
```

```python
df = pd.read_csv("../tests/data/raan.csv", parse_dates=True, index_col=0)

series = df['flow']

train = series.loc[:'2010']
test = series.loc['2020':]
```

```python
from tsod.detectors import SeasonalDetector

detector = SeasonalDetector(threshold=-5)

detector.fit(train)
```

```python
anom = detector.detect(test)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Scatter(go.Scatter(x=test.index, y=test, mode='markers+lines'), name="Flow"),
    go.Scatter(go.Scatter(x=test[anom].index, y=test[anom], mode='markers', name='Anomalies')),
    ], layout=go.Layout(width=1000))
fig.show() 
```

Introduce some outliers

```python
test['2020-07-01':'2020-07-06'] = [1.0, 2.0, 3.0,3.0, 2.0, 1.0]
test['2020-08-01':'2020-08-05'] = 2.0
```

```python
anom = detector.detect(test)
fig = go.Figure(data=[
    go.Scatter(go.Scatter(x=test.index, y=test, mode='markers+lines'), name="Flow"),
    go.Scatter(go.Scatter(x=test[anom].index, y=test[anom], mode='markers', name='Anomalies')),
    ], layout=go.Layout(width=1000))
fig.show()
```
